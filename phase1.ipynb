```python
def get_entity_rows(entity_name, *dataframes, target_count=None):
    rows = []
    for df in dataframes:
        if entity_name in df['entity_name'].values:
            filtered_df = df[df['entity_name'] == entity_name]
            rows.append(filtered_df)
    combined_df = pd.concat(rows, ignore_index=True)
    if target_count is not None:
        combined_df = combined_df.head(target_count)
    return combined_df
```

```python
# Get rows for each entity name
item_weight_df = get_entity_rows('item_weight', my6000, vish_3, ana_3, target_count=6000)
item_volume_df = get_entity_rows('item_volume', my6000, vish_3, ana_3, target_count=6000)
height_df = get_entity_rows('height', my6000, vish_4, ana_1, ana_3, target_count=6000)
maximum_weight_recommendation_df = get_entity_rows('maximum_weight_recommendation', vish_5, ana_3)
voltage_df = get_entity_rows('voltage', vish_5)
depth_df = get_entity_rows('depth', my6000, ana_1, ana_2)
wattage_df = get_entity_rows('wattage', my6000, vish_5)
width_df = get_entity_rows('width', vish_4, target_count=5000)
```

# Clean training data

```python
import pandas as pd
import re

def clean_text(text):
    # Remove unwanted symbols and extra spaces
    text = re.sub(r'[^\w\s\.\-\/]', '', text)  # Keep only alphanumeric, spaces, dots, dashes, and slashes
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = text.strip()  # Remove leading and trailing spaces
    return text

def clean_file(csvfilename, fname):
    # Load the dataset
    df = pd.read_csv(csvfilename)

    # Check if 'Extracted values' column exists
    if 'Extracted_text' not in df.columns:
        raise ValueError("Column 'Extracted_text' not found in the CSV file.")

    # Clean the 'Extracted values' column
    df['Extracted_text'] = df['Extracted_text'].apply(lambda x: clean_text(str(x)))

    # Save the cleaned dataframe to a new CSV file
    df.to_csv('training datasets/' + fname + '_cleaned.csv', index=False)
```

```python
clean_file('training datasets/item_weight.csv', 'item_weight')
clean_file('training datasets/maximum_weight_recommendation.csv', 'maximum_weight_recommendation')
clean_file('training datasets/item_volume.csv', 'item_volume')
clean_file('training datasets/height.csv', 'height')
clean_file('training datasets/voltage.csv', 'voltage')
clean_file('training datasets/depth.csv', 'depth')
clean_file('training datasets/width.csv', 'width')
clean_file('training datasets/wattage.csv', 'wattage')
```

```python
from datasets import Dataset

# Convert potential float values to bytes
shuffled_df['Extracted_text'] = shuffled_df['Extracted_text'].astype(str)
shuffled_df['entity_value'] = shuffled_df['entity_value'].astype(str)

data = {
    'input_text': shuffled_df['Extracted_text'].tolist(),
    'target_text': shuffled_df['entity_value'].tolist()
}

# Create a Dataset object
dataset = Dataset.from_dict(data)
```

```python
from sklearn.model_selection import train_test_split
# Split the dataset into training and validation sets
train_dataset, eval_dataset = train_test_split(dataset, test_size=0.1, shuffle=True)

# Create Dataset objects for train and eval
train_dataset = Dataset.from_dict(train_dataset)
eval_dataset = Dataset.from_dict(eval_dataset)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments
from transformers import DataCollatorForSeq2Seq

# Load the T5 model and tokenizer
model_name = 't5-small'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
# Tokenize the data
def preprocess_function(examples):
    inputs = examples['input_text']
    targets = examples['target_text']
    model_inputs = tokenizer(inputs, max_length=512, truncation=True)

    # Tokenize targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True)

    model_inputs['labels'] = labels['input_ids']
    return model_inputs
```

```python
tokenized_datasets = dataset.map(preprocess_function, batched=True)
tokenized_eval_dataset = eval_dataset.map(preprocess_function, batched=True)
```

```python
# Define training arguments
training_args = TrainingArguments(
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    output_dir='./results',
    evaluation_strategy="epoch",
    logging_dir='./logs',
    num_train_epochs=3,#3
    save_steps=10_000,
    save_total_limit=2,
    eval_steps=1_000,
)
```

```python
# Define a data collator
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_datasets,
    eval_dataset=tokenized_eval_dataset
)

# Train the model
trainer.train()
```

```python
# Save the fine-tuned model
model.save_pretrained('MyModel/fine_tuned_t5')
tokenizer.save_pretrained('MyModel/fine_tuned_t5')
```

```python
import torch
from transformers import BartTokenizer, BartForConditionalGeneration, Trainer, TrainingArguments
from datasets import Dataset
```

```python
# Load the training data
df = pd.read_csv("train.csv")  # Your training dataset containing 'extracted_text', 'entity_name', and 'entity_value'
```

```python
# Combine the extracted text and entity name as input, and entity value as the target
def preprocess_function(examples):
    return {
        'input_text': 'Extracted text: ' + examples['extracted_text'] + ' Entity name: ' + examples['entity_name'],
        'target_text': examples['entity_value']
    }
```

```python
# Apply preprocessing
df['input_text'] = df.apply(lambda row: 'Extracted text: ' + row['extracted_text'] + ' Entity name: ' + row['entity_name'], axis=1)
df['target_text'] = df['entity_value']

# Convert the DataFrame to a Hugging Face Dataset
dataset = Dataset.from_pandas(df[['input_text', 'target_text']])
```

```python


# Initialize the BART tokenizer and model
tokenizer = BartTokenizer.from_pretrained('facebook/bart-base')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-base')
```

```python


# Tokenize the dataset
def tokenize_function(examples):
    inputs = tokenizer(examples['input_text'], max_length=512, truncation=True, padding='max_length')
    targets = tokenizer(examples['target_text'], max_length=128, truncation=True, padding='max_length')

    inputs['labels'] = targets['input_ids']
    return inputs
```

```python
tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Split into train and validation sets (80% training, 20% validation)
train_test_split = tokenized_dataset.train_test_split(test_size=0.2)
train_dataset = train_test_split['train']
valid_dataset = train_test_split['test']
```

```python
# Define training arguments
training_args = TrainingArguments(
    output_dir="./bart_entity_value_prediction",
    evaluation_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    save_steps=500,
    save_total_limit=2
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=tokenizer
)

# Train the model
trainer.train()

# Save the final model
model.save_pretrained('./bart_entity_value_model')
tokenizer.save_pretrained('./bart_entity_value_model')
```
